```python
import kagglehub
from pathlib import Path
import shutil
```

```python
path = kagglehub.dataset_download("urvishp80/irmas-training-data")

destino = Path("datasets/irmas-training-data")
destino.parent.mkdir(parents=True, exist_ok=True)

if destino.exists():
    shutil.rmtree(destino)

shutil.copytree(path, destino)

print("Dataset baixado em cache:", path)
print("Dataset copiado para:", destino.resolve())
```
